```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X = np.array([
    [0, 1.5],
    [0, 2.5],
    [0, 3.5],
    [1, 1.5],
    [1, 3.5],
    [2, 1.5],
    [2, 2.5],
    [2, 2.5]
]) 
```

```python
y = np.array([2.3, 6.1, 1.3, 6.4, 2.6, 7.5, 5.8, 7.2])
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
a0 = model.intercept_
a1, a2 = model.coef_
```

```python
print(f"Рівняння: y = {a0:.2f} + {a1:.2f}*x1 + {a2:.2f}*x2")
```

```python
x1_test, x2_test = 1.5, 3
y_pred = model.predict(np.array([[x1_test, x2_test]]))
print(f"Передбачення y для x1={x1_test}, x2={x2_test}: {y_pred[0]:.2f}")
```

```python
r2 = r2_score(y, model.predict(X))
print(f"Коєфіцієнт детермінації (R^2): {r2:.4f}")
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:, 0], X[:, 1], y, color='red', label='Original Data')

x1_range = np.linspace(min(X[:, 0]), max(X[:, 0]), 10)
x2_range = np.linspace(min(X[:, 1]), max(X[:, 1]), 10)
X1, X2 = np.meshgrid(x1_range, x2_range)
Y_pred = a0 + a1 * X1 + a2 * X2

ax.plot_surface(X1, X2, Y_pred, alpha=0.5, cmap='viridis')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('Regression Plane')

plt.legend()
plt.show()
```
