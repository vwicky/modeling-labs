```python
import matplotlib.pyplot as plt
import networkx as nx

# Створення орієнтованого графа
G = nx.DiGraph()

# Вузли (стани) з мітками
nodes = {
    'M0': '[1, 0, 0, 0, 0]',
    'M1': '[0, 1, 1, 1, 0]',
    'M2': '[0, 0, 1, 1, 1]',
    'M3': '[0, 0, 0, 1, 2]',
    'M4': '[0, 0, 0, 0, 1]'
}

# Додавання вузлів
for node, label in nodes.items():
    G.add_node(node, label=label)

# Додавання ребер (перехідні дуги)
edges = [
    ('M0', 'M1', 't5'),
    ('M1', 'M2', 't2'),
    ('M2', 'M3', 't3'),
    ('M3', 'M4', 't4'),
    ('M4', 'M0', 't1')
]
G.add_edges_from([(src, dst, {'label': lbl}) for src, dst, lbl in edges])

# Побудова позицій для вузлів у колі
pos = nx.circular_layout(G)

# Побудова графа
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10)
labels = nx.get_node_attributes(G, 'label')
nx.draw_networkx_labels(G, pos, labels=labels, font_size=9, verticalalignment='bottom')
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9)

plt.title("Граф досяжності мережі Петрі")
plt.axis('off')
plt.tight_layout()
plt.savefig("/data/reachability_graph_matplotlib.png")
```
